## Song Data Analysis

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the data 
df = pd.read_csv('../data/complete_playlist_data.csv')
df.shape
```

```python
df.head()
```

```python
# Check the data types
df.dtypes
```

```python
# Convert the datetime format
df['release_date'] = pd.to_datetime(df['release_date'], dayfirst=True, errors = 'coerce')
df
```

```python
# Check the number of songs in each category 
df['decade'].value_counts().sort_index(ascending=True)
```

There is very little data for 1950s and 1940s so we can remove that from the data. 

```python
# Remove the decades 1950 and 1940 from the data 
df = df[(df['decade'] != 1940) & (df['decade'] != 1950)]
df.shape
```

```python
# Plot the decade wise song counts 
df['decade'].value_counts().sort_index(ascending=True).plot(kind='bar', title="Number of songs from each decade.", xlabel='year', ylabel='Number of songs')
```

```python
# Create a dataframe with the decade wise data
decade_wise_data = (
    df.groupby(["decade"])
    .agg(
        {
            "danceability": "mean",
            "acousticness": "mean",
            "energy": "mean",
            "loudness": "mean",
            "speechiness": "mean",
            "valence": "mean",
            "tempo": "mean",
            "time_signature": "mean",
            "mode": "mean",
            "length": "mean",
        }
    )
    .reset_index()
)
decade_wise_data
```

```python
# Save the decade wise charachterstics 
decade_wise_data.to_csv('../data/decade_wise_song_features.csv', index=False)
```

### Analyze key variations

```python
# Plot the boxplot for key and and all other features in subplots 
fig, axes = plt.subplots(3, 2, figsize=(10, 10))
sns.boxplot(x="key", y="danceability", data=df, ax=axes[0, 0])
sns.boxplot(x="key", y="acousticness", data=df, ax=axes[0, 1])
sns.boxplot(x="key", y="energy", data=df, ax=axes[1, 0])
sns.boxplot(x="key", y="loudness", data=df, ax=axes[1, 1])
sns.boxplot(x='key', y='valence', data=df, ax=axes[2, 0])
sns.boxplot(x='key', y='tempo', data=df, ax=axes[2, 1])
plt.show()
```

## Cluster Analysis

```python
# Import libraries 
from sklearn.metrics.pairwise import pairwise_distances
from scipy.spatial.distance import pdist, squareform
```

```python
# Create a similarity matrix using ecludian distance
similarity_matrix = pairwise_distances(decade_wise_data.iloc[:, 1:10], metric='euclidean')*10

# Plot the similarity matrix for averages
plt.figure(figsize=(12, 10))
sns.heatmap(similarity_matrix, 
            cmap='RdBu_r', 
            annot=True, 
            xticklabels=decade_wise_data['decade'], 
            yticklabels=decade_wise_data['decade'])
```

Lower the value, higher the similarity in the matrix. 

```python
# Create a similarity matrix using cosine similarity
similarity_matrix = pairwise_distances(decade_wise_data.iloc[:, 1:]*10, metric='cosine')*1000

# Plot the similarity matrix for averages
plt.figure(figsize=(12, 10))
sns.heatmap(similarity_matrix, 
            cmap='RdBu_r', 
            annot=True, 
            xticklabels=decade_wise_data['decade'], 
            yticklabels=decade_wise_data['decade'])
```

```python
# Create a similarity matrix using manhattan distance
similarity_matrix = pairwise_distances(decade_wise_data.iloc[:, 1:10]*10, metric='manhattan')

# Plot the similarity matrix for averages
plt.figure(figsize=(12, 10))
sns.heatmap(similarity_matrix, 
            cmap='RdBu_r', 
            annot=True, 
            xticklabels=decade_wise_data['decade'], 
            yticklabels=decade_wise_data['decade'])
```

### Build a UMAP for clustering

```python
# Import umap library
import umap.umap_ as umap

# Build the UMAP model for decade-wise data 
umap_reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, metric='cosine')
umap_embedding = umap_reducer.fit_transform(df.iloc[:, 8:17])

# Plot UMAP embedding
plt.figure(figsize=(12, 10))
sns.scatterplot(umap_embedding[:, 0], umap_embedding[:, 1], hue=df['decade'], cmap='RdBu_r')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('UMAP Embeddings for all songs.')
```
