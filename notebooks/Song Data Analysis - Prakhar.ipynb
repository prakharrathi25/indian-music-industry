## Song Data Analysis

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the data 
df = pd.read_csv('../data/complete_playlist_data.csv')
df.shape
```

```python
df.head()
```

```python
# Check the data types
df.dtypes
```

```python
# Convert the datetime format
df['release_date'] = pd.to_datetime(df['release_date'], dayfirst=True, errors = 'coerce')
df
```

```python
# Check the number of songs in each category 
df['decade'].value_counts().sort_index(ascending=True)
```

There is very little data for 1950s and 1940s so we can remove that from the data. 

```python
# Remove the decades 1950 and 1940 from the data 
df = df[(df['decade'] != 1940) & (df['decade'] != 1950)]
df.shape
```

```python
# Plot the decade wise song counts 
df['decade'].value_counts().sort_index(ascending=True).plot(kind='bar', )
```

```python
# Create a dataframe with the decade wise data
decade_wise_data = (
    df.groupby(["decade"])
    .agg(
        {
            "danceability": "mean",
            "acousticness": "mean",
            "energy": "mean",
            "loudness": "mean",
            "speechiness": "mean",
            "valence": "mean",
            "tempo": "mean",
            "time_signature": "mean",
            "mode": "mean",
            "length": "mean",
        }
    )
    .reset_index()
)
decade_wise_data
```

```python
# Save the decade wise charachterstics 
decade_wise_data.to_csv('../data/decade_wise_song_features.csv', index=False)
```

## Cluster Analysis

```python
# Import libraries 
from sklearn.metrics.pairwise import pairwise_distances
from scipy.spatial.distance import pdist, squareform
```

```python
# Create a similarity matrix using ecludian distance
similarity_matrix = pairwise_distances(decade_wise_data.iloc[:, 1:], metric='euclidean')

# Plot the similarity matrix for averages
plt.figure(figsize=(12, 10))
sns.heatmap(similarity_matrix, 
            cmap='RdBu_r', 
            annot=True, 
            xticklabels=decade_wise_data['decade'], 
            yticklabels=decade_wise_data['decade'])
```

Lower the value, higher the similarity in the matrix. 

```python
# Create a similarity matrix using cosine similarity
similarity_matrix = pairwise_distances(decade_wise_data.iloc[:, 1:], metric='cosine')*1000

# Plot the similarity matrix for averages
plt.figure(figsize=(12, 10))
sns.heatmap(similarity_matrix, 
            cmap='RdBu_r', 
            annot=True, 
            xticklabels=decade_wise_data['decade'], 
            yticklabels=decade_wise_data['decade'])
```

```python
# Create a similarity matrix using manhattan distance
similarity_matrix = pairwise_distances(decade_wise_data.iloc[:, 1:], metric='manhattan')

# Plot the similarity matrix for averages
plt.figure(figsize=(12, 10))
sns.heatmap(similarity_matrix, 
            cmap='RdBu_r', 
            annot=True, 
            xticklabels=decade_wise_data['decade'], 
            yticklabels=decade_wise_data['decade'])
```
